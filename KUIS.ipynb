```python
print(f'1.')

# Import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data jumlah pelanggan (X) dan pendapatan (Y)
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)  # ubah jadi bentuk kolom
Y = np.array([200, 400, 600, 800, 1000])

# Buat model regresi linear
model = LinearRegression()
model.fit(X, Y)

# Cetak koefisien dan intercept
print(f'Intercept (b0): {model.intercept_}')
print(f'Koefisien (b1): {model.coef_[0]}')

# Prediksi pendapatan jika jumlah pelanggan = 35
X_pred = np.array([[35]])
Y_pred = model.predict(X_pred)
print(f'Prediksi pendapatan untuk 35 pelanggan: {Y_pred[0]} ribu rupiah')

# Visualisasi data dan garis regresi
plt.scatter(X, Y, color='blue', label='Data Asli')
plt.plot(X, model.predict(X), color='red', label='Garis Regresi')
plt.scatter(35, Y_pred, color='green', label='Prediksi (35 pelanggan)')
plt.title('Regresi Linear Jumlah Pelanggan vs Pendapatan')
plt.xlabel('Jumlah Pelanggan')
plt.ylabel('Pendapatan (ribu rupiah)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f'2.')
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data
X = np.array([1, 2, 3, 4]).reshape(-1, 1)
Y = np.array([60, 65, 70, 75])

# Buat model regresi linear
model = LinearRegression()
model.fit(X, Y)

# Ambil koefisien sebagai rata-rata peningkatan per jam belajar
rata_rata_peningkatan = model.coef_[0]
print(f'Rata-rata peningkatan nilai per jam belajar (dengan regresi): {rata_rata_peningkatan}')

# Visualisasi
plt.scatter(X, Y, color='blue', label='Data Asli')
plt.plot(X, model.predict(X), color='red', label='Garis Regresi')
plt.title('Waktu Belajar vs Nilai Ujian')
plt.xlabel('Jam Belajar')
plt.ylabel('Nilai Ujian')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f'3.')
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data Usia (X) dan Harga Jual (Y)
usia = np.array([1, 3, 5, 7, 9]).reshape(-1, 1)
harga = np.array([150, 120, 90, 60, 30])  # dalam juta rupiah

# Buat model regresi linear
model = LinearRegression()
model.fit(usia, harga)

# Ambil koefisien dan intercept
b = model.coef_[0]   # slope (kemiringan)
a = model.intercept_  # intercept (titik potong)

print(f"Persamaan regresi: Y = {a:.2f} + {b:.2f}X")

# Prediksi harga jual jika usia kendaraan 4 tahun
usia_prediksi = np.array([[4]])
harga_prediksi = model.predict(usia_prediksi)
print(f"Prediksi harga jual jika usia kendaraan 4 tahun: {harga_prediksi[0]:.2f} juta rupiah")

# Visualisasi
plt.scatter(usia, harga, color='blue', label='Data Asli')
plt.plot(usia, model.predict(usia), color='red', label='Garis Regresi')
plt.scatter(4, harga_prediksi, color='green', label='Prediksi (usia 4 thn)')
plt.xlabel("Usia Kendaraan (tahun)")
plt.ylabel("Harga Jual (juta rupiah)")
plt.title("Regresi Linear: Usia Kendaraan vs Harga Jual")
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f'4.')
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data Jam Internet (X) dan Biaya Pulsa (Y)
jam_internet = np.array([5, 10, 15, 20]).reshape(-1, 1)
biaya_pulsa = np.array([25000, 50000, 75000, 100000])  # dalam rupiah

# Buat model regresi linear
model = LinearRegression()
model.fit(jam_internet, biaya_pulsa)

# Ambil koefisien dan intercept
b = model.coef_[0]
a = model.intercept_

print(f"Persamaan regresi: Y = {a:.2f} + {b:.2f}X")

# Prediksi biaya pulsa untuk 18 jam
jam_prediksi = np.array([[18]])
pulsa_prediksi = model.predict(jam_prediksi)
print(f"Prediksi biaya pulsa untuk 18 jam penggunaan internet: Rp {pulsa_prediksi[0]:,.0f}")

# Visualisasi
plt.scatter(jam_internet, biaya_pulsa, color='blue', label='Data Asli')
plt.plot(jam_internet, model.predict(jam_internet), color='red', label='Garis Regresi')
plt.scatter(18, pulsa_prediksi, color='green', label='Prediksi (18 jam)')
plt.xlabel("Jam Internet")
plt.ylabel("Biaya Pulsa (Rp)")
plt.title("Regresi Linear: Jam Internet vs Biaya Pulsa")
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f'5.')
import numpy as np
from sklearn.linear_model import LinearRegression

# Data input
X1 = [50, 60, 70, 80, 90]  # Pengunjung
X2 = [2, 3, 4, 5, 6]       # Menu Baru
Y = [400, 480, 560, 640, 720]  # Penjualan (dalam ribu)

# Gabungkan X1 dan X2 jadi satu matriks X
X = np.column_stack((X1, X2))

# Buat model regresi linear berganda
model = LinearRegression()
model.fit(X, Y)

# Ambil koefisien dan intercept
a = model.intercept_
b1, b2 = model.coef_

print(f"Model regresi: Y = {a:.2f} + {b1:.2f}*X1 + {b2:.2f}*X2")

# Prediksi penjualan jika pengunjung = 75 dan menu baru = 4
X_pred = np.array([[75, 4]])
y_pred = model.predict(X_pred)

print(f"Prediksi penjualan untuk 75 pengunjung dan 4 menu baru: {y_pred[0]:,.2f} ribu rupiah")
```
